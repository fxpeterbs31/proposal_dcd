### Load Data

Pada library **`spotipy`**, telah terdapat fungsi **`current_user_top_tracks`**. Fungsi ini digunakan untuk **mendapatkan lagu-lagu favorit kita dalam suatu periode**. 

Ada **3 parameter** yang digunakan pada fungsi ini, yaitu: 
1. **`limit`** <br>
    **`limit`** merupakan **jumlah data yang ingin diambil**. **Default** dari parameter ini adalah **20**. Pada project ini, kita akan menggunakan **`limit = 50`** karena **kita ingin mengambil 50 lagu teratas**. <br><br>
2. **`offset`**<br>
    **`offset`** merupakan **peringkat awal lagu**. **Default** dari parameter ini adalah **0**. Pada project ini, kita menggunakan **`offset = 0`**. Dengan `limit = 50`, ini berarti **lagu yang diambil adalah lagu-lagu favorit kita dengan peringkat 1 - 50** <br><br>
3. **`time_range`** <br>
    **`time_range`** merupakan **periode waktu**. **Default** dari parameter ini adalah **`medium_term`**. Pada project ini, kita menggunakan **`time_range = 'short_term'`**, karena **periode waktu yang digunakan pada project ini adalah 4 minggu / 1 bulan terakhir**. Berikut detail dari value parameter ini. 
    - **`short_term`**<br>
        Periode waktu yang digunakan adalah **4 minggu / 1 bulan terakhir** <br><br>
    - **`medium_term`**<br>
        Periode waktu yang digunakan adalah **6 bulan terakhir**<br><br>
    - **`long_term`**<br>
        Periode waktu yang digunakan adalah **seluruh data / all time** <br><br>

**Output** dari fungsi ini kita **simpan** kedalam sebuah **variabel** yang bernama **`top_50_4_week`**

```python
top_50_4_week = sp.current_user_top_tracks(limit = 50, time_range = 'short_term')
```

```python
li_uri               = []
li_name              = []
li_artist_uri        = []
li_artist_name       = []
li_artist_popularity = []
li_artist_genres     = []
li_album             = []
li_track_popularity  = []
li_acousticness      = []
li_danceability      = []
li_duration_ms       = []
li_energy            = []
li_instrumentalness  = []
li_key               = []
li_liveness          = []
li_loudness          = []
li_mode              = []
li_speechiness       = []
li_tempo             = []
li_time_signature    = []
li_valence           = []

for song in top_50_4_week['items']: 
    song_id = song['id']
    
    uri = song['uri']
    li_uri.append(uri)
    
    name = song['name']
    li_name.append(name)
    
    artist_uri = song['artists'][0]['uri']
    li_artist_uri.append(artist_uri)
    
    artist_name = song['artists'][0]['name']
    li_artist_name.append(artist_name)
    
    artist_info = sp.artist(artist_uri)
    
    artist_popularity = artist_info['popularity']
    li_artist_popularity.append(artist_popularity)
    
    artist_genres = artist_info['genres']
    li_artist_genres.append(artist_genres)
    
    album = song['album']['name']
    li_album.append(album)
    
    track_popularity = song['popularity']
    li_track_popularity.append(track_popularity)
    
    audio_features = sp.audio_features(song_id)[0]
    
    acousticness = audio_features['acousticness']
    li_acousticness.append(acousticness)      

    danceability = audio_features['danceability']
    li_danceability.append(danceability)      

    duration_ms = audio_features['duration_ms']
    li_duration_ms.append(duration_ms)       

    energy = audio_features['energy']
    li_energy.append(energy)            

    instrumentalness = audio_features['instrumentalness']
    li_instrumentalness.append(instrumentalness)  

    key = audio_features['key']
    li_key.append(key)               

    liveness = audio_features['liveness']
    li_liveness.append(liveness)          

    loudness = audio_features['loudness']
    li_loudness.append(loudness)          

    mode = audio_features['mode']
    li_mode.append(mode)              

    speechiness = audio_features['speechiness']
    li_speechiness.append(speechiness)       

    tempo = audio_features['tempo']
    li_tempo.append(tempo)             

    time_signature = audio_features['time_signature']
    li_time_signature.append(time_signature)    

    valence = audio_features['valence']
    li_valence.append(valence)           
```

### Converting To Dataframe

```python
df_4_week = pd.DataFrame(data = {
    'name' : li_name,
    'artist_name' : li_artist_name,
    'artist_popularity' : li_artist_popularity,
    'artist_genres' : li_artist_genres,
    'album_name' : li_album,
    'track_popularity' : li_track_popularity,
    'acousticness' : li_acousticness,
    'danceability' : li_danceability,
    'duration_ms' : li_duration_ms,
    'energy' : li_energy,
    'instrumentalness' : li_instrumentalness,
    'key' : li_key,
    'liveness' : li_liveness,
    'loudness' : li_loudness,
    'mode' : li_mode,
    'speechiness' : li_speechiness,
    'tempo' : li_tempo,
    'time_signature' : li_time_signature,
    'valence' : li_valence
})
```

### Formatting Data

#### Splitting `artist_genres`

Pada **`df_4_week`**, bisa dilihat pada kolom **`artist_genres`** bahwa **value** dari kolom tersebut berupa sebuah **list** yang **memiliki 1 atau lebih value**. Hal seperti ini **ingin kita hindari**, karena **terlihat seperti tumpang tindih**. Untuk mengatasi hal ini, saya akan **membuat kolom baru** yang bernama **`artist_genres_split`** yang berisi **`artist_genres`** yang sudah **di-split**, dan **setiap baris** hanya **memiliki 1 value**. Setelah kolom `artist_genres_split` terisi, saya akan **buang** kolom **`artist_genres`**.

```python
len_genres = []
li_genres = []

for i in range(df_4_week.shape[0]):    
    genres = df_4_week.loc[i, 'artist_genres']
    ln = len(genres)
    
    for j in range(ln):
        li_genres.append(genres[j])
    
    len_genres.append(ln)
```

```python
df_4_week['len_genres'] = len_genres
```

```python
df_4_week = df_4_week.loc[df_4_week.index.repeat(df_4_week['len_genres'])].reset_index(drop = True)
```

```python
df_4_week['artist_genres_split'] = li_genres
```

```python
df_4_week = df_4_week.drop(['artist_genres', 'len_genres'], axis = 1)
```

#### Replace `mode` values

Pada kolom **`mode`**, hanya terdapat **2** value yaitu **0** dan **1**. **0** menandakan bahwa **lagu tersebut menggunakan tangga nada minor**, sedangkan **1** menandakan bahwa **lagu tersebut menggunakan tangga nada major**. Value dari kolom **`mode`** akan **lebih bisa merepresentasikan value** mereka **apabila** **0** dan **1** **diubah** menjadi **Minor** dan **Major**. Hal ini bisa kita lakukan dengan menggunakan fungsi **`replace()`**

```python
df_4_week['mode'] = df_4_week['mode'].replace({
    0 : 'Minor',
    1 : 'Major'
})
```

#### Replace `key` values

Pada kolom **`key`**, hanya terdapat **12** value yaitu **0** sampai **11** dan berupa **integer**. **0** menandakan bahwa **lagu tersebut menggunakan nada dasar do = C**, **1** menandakan bahwa **lagu tersebut menggunakan nada dasar do = C♯/D♭**, dan seterusnya sampai **11** menyatakan bahwa **lagu tersebut menggunakan nada dasar do = B**. Value dari kolom **`key`** akan **lebih bisa merepresentasikan value** mereka **apabila** **0**, **1**, dst.  **diubah** menjadi **C**, **C♯/D♭**, dst. . Hal ini bisa kita lakukan dengan menggunakan fungsi **`replace()`**

```python
df_4_week['key'] = df_4_week['key'].replace({
    0  : 'C',
    1  : 'C♯/D♭',
    2  : 'D',
    3  : 'D♯/E♭',
    4  : 'E',
    5  : 'F',
    6  : 'F♯/G♭',
    7  : 'G',
    8  : 'G♯/A♭',
    9  : 'A',
    10 : 'A♯/B♭',
    11 : 'B'
})
```

```python
df_4_week.head()
```

#### Replace `time_signature` values

Pada kolom **`time_signature`**, hanya terdapat **5** value yaitu **3** sampai **7** dan berupa **integer**. **3** menandakan bahwa **lagu tersebut menggunakan birama 3/4**, **4** menandakan bahwa **lagu tersebut menggunakan birama 4/4**, dan seterusnya sampai **7** menyatakan bahwa **lagu tersebut menggunakan birama 7/4**. Value dari kolom **`time_signature`** akan **lebih bisa merepresentasikan value** mereka **apabila** **3**, **4**, dst.  **diubah** menjadi **3/4**, **4/4**, dst. . Hal ini bisa kita lakukan dengan menggunakan fungsi **`replace()`**

```python
df_4_week['time_signature'] = df_4_week['time_signature'].replace({
    3 : '3/4',
    4 : '4/4',
    5 : '5/4',
    6 : '6/4',
    7 : '7/4'
})
```
